```python
import os
import sys
from obspy.io.segy.core import _read_segy
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib
from PIL import Image
import math
from scipy.misc import toimage,imsave
%matplotlib inline

def img2d_plot(array,figsize = (10,10),colormap = "Greys",gain = -2):
    img = array
    crange = np.sqrt(np.mean(np.square(img)))
    vmin = - math.exp(-gain) * crange
    vmax = -vmin
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    ax.imshow(img.T,cmap = colormap,vmin = vmin, vmax = vmax)
    fig.show()
    plt.savefig("tem.jpg")
    return
def img2d_save(array,im_name):
    im = toimage(array.T)
    im.save(im_name +".jpg")
```

```python
## QC sections
imgs = os.listdir("../data/Teapot_dome/")
imgname = np.random.choice(imgs)
print(imgname)
imgtst = np.loadtxt("../data/Teapot_dome/" + imgname)
img2d_plot(imgtst,figsize = (5,5),gain = 0)
```

```python
## Save all the images as jpg
for item in imgs:
    imgtst = np.loadtxt("../data/Teapot_dome/" + item)
    img2d_save(imgtst,"../data/Seismic/" + item)
```
